## 課題1

各問で示されている図を出力してください。

その際、以下の要件を必ず満たすようにしてください。


1.   for文を必ず用いること
2.   変数`size`の値を変えることで自由に図形の大きさを変更できること




### 問1

以下の図形を出力してください。

```
*
**
***
****
*****
```

````python
size = 5
# 問1: 直角三角形パターン
for i in range(1, size + 1):
    print('*' * i)
````

### 問2

以下の図形を出力してください。

```
*
 *
  *
   *
    *
```

````python
size = 5
# 問2: 対角線パターン（右下がり）
for i in range(size):
    print(' ' * i + '*')
````

### 問3

以下の図形を出力してください。

```
    *
   *
  *
 *
*
```

````python
size = 5
# 問3: 対角線パターン（右上がり）
for i in range(size):
    print(' ' * (size - 1 - i) + '*')
````

### 問4

以下の図形を出力してください。

```
*   *
 * *
  *
 * *
*   *
```

````python
size = 5
# 問4: Xパターン
for i in range(size):
    # 上半分と下半分で処理を分ける
    if i < size // 2:
        # 上半分：外側から内側へ
        spaces_left = i
        spaces_middle = (size - 2 * i - 2)
        print(' ' * spaces_left + '*' + ' ' * spaces_middle + '*')
    elif i == size // 2:
        # 中央：1つの*のみ
        print(' ' * (size // 2) + '*')
    else:
        # 下半分：内側から外側へ
        spaces_left = size - 1 - i
        spaces_middle = (2 * i - size)
        print(' ' * spaces_left + '*' + ' ' * spaces_middle + '*')
````

### 問5

以下の図形を出力してください。

```
    *
   ***
  *****
 *******
*********
```

````python
size = 5
# 問5: ピラミッドパターン
for i in range(size):
    # 左側のスペース数
    spaces = size - 1 - i
    # 星の数（奇数）
    stars = 2 * i + 1
    print(' ' * spaces + '*' * stars)
````
