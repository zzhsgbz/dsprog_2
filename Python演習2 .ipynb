## 課題1

1.   for文を必ず用いること
2.   変数`size`の値を変えることで自由に図形の大きさを変更できること




### 問1

以下の図形を出力してください。

```
*
**
***
****
*****
```

````python
size = 5

for i in range(1, size + 1):
    print('*' * i)
````

### 問2

以下の図形を出力してください。

```
*
 *
  *
   *
    *
```

````python
size = 5
for i in range(size):
    print(' ' * i + '*')
````

### 問3

以下の図形を出力してください。

```
    *
   *
  *
 *
*
```

````python
size = 5

for i in range(size):
    print(' ' * (size - 1 - i) + '*')
````

### 問4

以下の図形を出力してください。

```
*   *
 * *
  *
 * *
*   *
```

````python
size = 5
for i in range(size):
    pos1 = i  
    pos2 = size - 1 - i  
    
    if pos1 == pos2: 
        print(' ' * pos1 + '*')
    else:
        left = min(pos1, pos2)
        right = max(pos1, pos2)
        print(' ' * left + '*' + ' ' * (right - left - 1) + '*')
````

### 問5

以下の図形を出力してください。

```
    *
   ***
  *****
 *******
*********
```

````python
size = 5
for i in range(size):
    spaces = size - 1 - i
    stars = 2 * i + 1
    print(' ' * spaces + '*' * stars)
````
